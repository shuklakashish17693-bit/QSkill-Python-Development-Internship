```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO

# Sample CSV data create kar rahe hain
data = """Name,Age,Salary,Department
Alice,25,50000,IT
Bob,30,60000,HR
Charlie,35,75000,IT
Diana,28,55000,Finance
Eve,32,80000,IT
Frank,27,45000,HR
Grace,31,70000,Finance
Henry,29,65000,IT"""

# CSV load karo
df = pd.read_csv(StringIO(data))
print("Data loaded successfully!")
print(df)
print("\nAverage Salary:", df['Salary'].mean())
print("Average Age:", df['Age'].mean())
```

```python
# First, we need to create or load the DataFrame
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.DataFrame({
    'Department': ['IT', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR'],
    'Age': [25, 30, 35, 28, 32, 40, 26, 29],
    'Salary': [70000, 50000, 60000, 75000, 55000, 80000, 72000, 52000]
})


# 1. Bar Chart - Department wise Salary
plt.figure(figsize=(8,5))
df.groupby('Department')['Salary'].mean().plot(kind='bar', color=['blue','green','orange'])
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.xticks(rotation=0)
plt.show()

# 2. Scatter Plot - Age vs Salary
plt.figure(figsize=(8,5))
plt.scatter(df['Age'], df['Salary'], color='red', s=100)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()

# 3. Heatmap - Correlation
plt.figure(figsize=(6,4))
sns.heatmap(df[['Age','Salary']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

print("Insights:")
print("1. IT department has highest average salary")
print("2. Salary increases with age generally")
print("3. Age and Salary are positively correlated")
```

```python
# Final Analysis and Observations
print("=" * 50)
print("DATA ANALYSIS REPORT")
print("=" * 50)

print("\n📊 BASIC STATISTICS:")
print(df.describe())

print("\n📈 DEPARTMENT WISE ANALYSIS:")
dept_analysis = df.groupby('Department').agg({
    'Salary': ['mean', 'max', 'min'],
    'Age': 'mean'
}).round(2)
print(dept_analysis)

print("\n💡 KEY INSIGHTS:")
print("1. Total Employees:", len(df))
print("2. Average Salary: ₹", df['Salary'].mean())
print("3. Highest Salary: ₹", df['Salary'].max())
print("4. Lowest Salary: ₹", df['Salary'].min())
print("5. Most Common Department:", df['Department'].mode()[0])
print("6. IT department employees earn the most on average")
print("7. Finance department has second highest salaries")
print("8. HR department has the lowest average salary")
```
